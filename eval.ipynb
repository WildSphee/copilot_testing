Import Necessary Modules

```python
import glob
import os
from datetime import datetime

import pandas as pd

# Import the functions from your scripts
from data_prep import generate_copilot_answers
from eval.ragas_eval import evaluate_with_ragas
```

Ensure you have ur .env set up according to README.md

```python
# Define the path to your input CSV file (testing_set.csv)
input_csv_path = "testing_set.csv"  # Update the path if the file is located elsewhere

# Define the directory where the copilot results will be saved
output_csv_dir = "results"  # You can change this to your preferred directory

# Ensure the output directory exists
os.makedirs(output_csv_dir, exist_ok=True)
```

Generate Context: using all files within the `context/` folder

```python
from data_prep import generate_context

context = generate_context()
```

# Generate copilot answers

```python
generate_copilot_answers(input_csv_path, output_csv_dir)
# Get the latest copilot_result CSV file from the output directory
list_of_files = glob.glob(f"{output_csv_dir}/copilot_result_*.csv")
latest_file = max(list_of_files, key=os.path.getctime)

print(f"Latest copilot result file: {latest_file}")
```

Evaluate with RAGAS

```python
# Define the output path for the evaluation results
evaluation_output_csv = (
    f"{output_csv_dir}/evaluation_results_{datetime.now().strftime('%m_%d_%H%M')}.csv"
)

# Evaluate using RAGAS
evaluate_with_ragas("results/testing.csv", evaluation_output_csv, context)
```

View Evaluation Results
Let's load and view the evaluation results.

```python
# Load the evaluation results
evaluation_df = pd.read_csv(evaluation_output_csv)

# Display the first few rows
evaluation_df.head()
```
